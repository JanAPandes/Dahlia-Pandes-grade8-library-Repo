```python
import requests
import json
url = "https://raw.githubusercontent.com/JanAPandes/Dahlia-Pandes-grade8-library-Repo/refs/heads/main/books.json"
response = requests.get(url)
books = response.json()
print(books)
```

```python
#Display book titles.

count = 0
for s in books:
  for s in books["books"]:
    arr = books["books"][count]["title"]
    print(arr)
    count += 1
```

```python
#Display title and availability
count = 0
for s in books:
  for s in books["books"]:
    title = books["books"][count]["title"]
    availability = books["books"][count]["details"]["available"]
    print(f"Title: {title}")
    print(f"Availability: {availability} \n")
    count += 1
```

```python
#Total number of books in library.

count = 0
sum = 0
for s in books:
  for s in books["books"]:
    copies = books["books"][count]["details"]["copies"]
    sum += copies
    count += 1

print(f"There are a total of {sum} books in the library.")
```
